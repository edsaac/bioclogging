# Export to C++ code

1. Load the PKL file containing the Fourier information
2. Write a mapper for X-coordinate
2. Write each (real,imag) pair

Structure: 
```
RealValues[xcoord][frequencies] = {...};
ImagValues[xcoord][frequencies] = {...};
```

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pickle, random
from os import system
````

````python
# Load data
RAWFILE = "pickledData_constg/PTX(4).pkl"
with open(RAWFILE,'rb') as f: data = pickle.load(f)
````

````python
data
````

````python
# Scale Fourier magnitudes
## The lenght of the original fourier series
scaleFactor = 9351 
data.iloc[0] = data.iloc[0]/2.0  ## DC is doubled (rfft)
data /= scaleFactor              ## Scale the whole thing
````

````python
# Split and transpose for C++ efficiency
real = np.real(data).T
imag = np.imag(data).T
````

````python
# Write files
folder = "cppCodebits_constg/"
fpaths = [folder+f"{i}Coefs.csv" for i in ["real","imag"]]
save_dict = dict(fmt='%12.6E',delimiter=",",newline=" \n ",\
                 header="",footer="",comments='')

for f,vals in zip(fpaths,[real,imag]) : np.savetxt(f,vals,**save_dict)
````

````python
# Final touches
for f in fpaths : 
    system(f"sed -i 's/^/{{ /g' {f}")
    system(f"sed -i 's/$/ }},/g' {f}")
    system(f"sed -i '1i{{ ' {f}")
    system(f"sed -i '$a }};' {f}")
````

#### Remember to clean the output CSV to be read from the C++ code
i.e., erase some curly braces and commas at the header and end of the files.

These files are then copy/pasted in the C++ code as array initializations.
