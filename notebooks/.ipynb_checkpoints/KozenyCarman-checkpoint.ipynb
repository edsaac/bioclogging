```python
%reset -f
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gs

from ipywidgets import interact_manual
import ipywidgets as widgets
```

```python
rho = 0.9970470    #g/cm3
g   = 981.0  #cm/s2
mu  = 0.0089 #g/mÂ·s
d   = 0.042  #cm

Porosity = np.linspace(0.001,0.999,100)
K = d*d/180.0 * (Porosity**3)/((1-Porosity)**2) * rho * g / mu
```

```python
PorositySand = 0.48
KSand = d*d/180.0 * (PorositySand**3)/((1-PorositySand)**2) * rho * g / mu
print(KSand) 
print(np.log10(KSand))
```

```python
fig = plt.figure(figsize=(6,10),facecolor="white");
ax1 = plt.subplot(2,1,1)
ax1.plot(Porosity,K,lw=3,c="green")

ax1.set_yscale('log')
ax1.set_xlim(0.05,0.95)
ax1.set_ylim(1.0E-4,1.0E4)
ax1.set_xlabel("Porosity $\\theta$ ")
ax1.set_ylabel("Hydraulic conductivity \n$K$ [cm/s]")
ax1.axvline(x=PorositySand,ls="dotted",lw=2,c="gray")
ax1.annotate("$K = %.3f$ cm/s"  \
         %(KSand),\
         xy=(PorositySand, 1.0E-3),\
         ha='center',bbox=dict(facecolor='springgreen', alpha=0.5))
plt.show()
```
